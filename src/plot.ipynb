```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 1
path = f"data/ngrams_{N}/"

test = pd.read_csv(path + 'test_result.txt', sep='    ')
```

```python
test['origin'] = test.index
#reset index
test = test.reset_index(drop=True) 
```

```python
# barplot
melted = pd.melt(test, value_vars=[ 'precision', 'recall', 'f1-score'], id_vars=['origin'])
melted
```

```python
sns.set_theme (style="whitegrid", font_scale=1.5,palette='Dark2')

fig, ax = plt.subplots(figsize=(10, 6), )
sns.barplot(x='origin', y='value', hue='variable', data=melted, ax=ax)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.set_title(f"ngram = {N}")

#rotate xtick
x = plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
```
